# Summer Olympics Data Analysis Assignment

###IMPORTING THE NECESSARY LIABRARIES AND THE DATASET CSV FILE

```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Summer_Olympics.csv")
```

### 1. In how many cities Summer Olympics is held so far?

```python
unique_cities=[]
cities=len(df['City'].unique())
unique_cities.append(df['City'].unique())
print("Summer Olympics is held in ", cities, "cities across the world")
for cities in unique_cities:
    print("\n These are the cities in which olympics were held : \n\n ", cities)
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
Top_5_sports=df[df['Medal']=='Gold']['Sport'].value_counts().head()
print("The top 5 sports in which the GOLD MEDALS WERE GIVEN : \n", Top_5_sports)
print("\n THE GRAPH IS SHOWN BELOW: ")
Top_5_sports.plot(kind="bar", color="Blue", figsize=(12,5))
plt.xlabel("Sports ")
plt.ylabel("Number of Gold Medals Won in these sports")
plt.title("The Top 5 Sports in which Max Gold medals ")
plt.show()
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
Top_5_sports=df['Sport'].value_counts().head()
print("The Sports that are having the most NUmber of MEDALS GOT : are \n",Top_5_sports)
print("\n THE GRAPH IS SHOWN BELOW: ")
Top_5_sports.plot(kind="bar", color="Orange", figsize=(12,5))
plt.xlabel("Sports")
plt.ylabel("Number of medals Won in total")
plt.title("TOP 5 sports in which most number of medals were won")
plt.show()
```

### 4. Which player has won most number of medals? (Top 5)

```python
Top_5_Players=df['Athlete'].value_counts().head()
print("The Top 5 Players that won the most number of medals are : \n",Top_5_Players)
print("\n THE GRAPH IS SHOWN BELOW: ")
Top_5_Players.plot(kind="bar", color="green", figsize=(12,5))
plt.xlabel("Players names who WON THE MOST MEDALS")
plt.ylabel("Number of medals Won by that player")
plt.title(" The TOP 5 players who won the most number of  medals")
plt.show()
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
Top_5_Players=df[df['Medal']=='Gold']['Athlete'].value_counts().head()
print("The player who has WON the most number of Gold Medals are : \n ",Top_5_Players)
print("\n THE GRAPH IS SHOWN BELOW: ")
Top_5_Players.plot(kind="bar", color="Red", figsize=(12,6))
plt.xlabel("Players names who WON THE MOST number of GOLD MEDALS")
plt.ylabel("Number of Gold medals Won by that player")
plt.ylim(0,18,1)
plt.title(" The TOP 5 players who won the most number of  medals")
plt.show()
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
data=df[df['Country']=='IND']
result = data[data['Medal']== 'Gold']['Year'].sort_values(ascending = True).head(1).to_string(index = False)

print("\n India won its first summer olympic gold in the year '", result,"'." )
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
Most_popular_event=df['Event'].value_counts().head()
print("The most Popular Event in terms of NUmber of Players are : \n", Most_popular_event)
print("\n THE GRAPH IS SHOWN BELOW: ")
Most_popular_event.plot(kind="bar", color="Purple", figsize=(12,6))
plt.xlabel("Most Popular Events ")
plt.ylabel("Number of Athelets who plays this event")
plt.title(" The Most popular events in terms of NUmber of ATHELETS ")
plt.show()
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
females=df[df['Gender']=='Women']
females=females[females['Medal']=='Gold']
Females_Gold_medalists=females['Sport'].value_counts().head()
print(" The top 5 Sports having the most number of Gold Medalists are : \n", Females_Gold_medalists)
print("\n THE GRAPH IS SHOWN BELOW: ")
Females_Gold_medalists.plot(kind="bar", color="Pink", figsize=(12,6))
plt.xlabel("The Sports having the most number of FEMALE GOLD MEDALISTS ")
plt.ylabel("Number of Females Athelets who Won GOLD in these events")
plt.title(" The Most popular Sports in which FEMALES have won the MOST GOLD MEDALS")
plt.show()
```
